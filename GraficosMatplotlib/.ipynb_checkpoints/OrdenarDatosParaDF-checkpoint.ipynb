### DataFrame Disac cuya unidad de medida está en Kg. 

```python
df_disac_kgr = df_disac[df_disac["IDMEDIDA"]== 'KGR ']
df_disac_kgr.head(0)
```

### DataFrame Disac cuya unidad de medida está en quintales(QNTL)

```python
df_disac_qntl = df_disac[df_disac["IDMEDIDA"]== 'QNTL']
df_disac_qntl.head(5)
```

#### Hallar peso neto total por kg y qntl 

```python
peso_kgr = df_disac_kgr["PESONETO"].sum() 
peso_qntl = df_disac_qntl["PESONETO"].sum() 

print('Peso neto en KGR: {} kg'.format(peso_kgr))
print('Peso neto en QNTL: {} qn'.format(peso_qntl))
```

### Hallar peso neto total Año:2014,2015,2016 
#### Consideraciones: Su equivalente de 1QNTL es 46,02KGR

```python
pneto_total = peso_kgr + (peso_qntl * 46.02)
print('Peso neto total de todos los registros: {} kg'.format(pneto_total))
```

#### Función para hallar los peso por año de cierta unidad de medida 

```python
def get_sumpeso_anio(df_m, anio_str):
    suma_anio_um = 0
    for i in range(0, len(df_m)):
        anio = str( df_m.get_value(df_m.index[i],'FECHACREACION')).split("-")[0]
        if anio == anio_str :
            suma_anio_um = suma_anio_um + df_m.get_value(df_m.index[i],'PESONETO')
    return suma_anio_um
```

#### Peso neto total en kg para el año 2014

```python
# Suma Peso de Kg para 2014 
suma_2014_kg = get_sumpeso_anio(df_disac_kgr, '2014')
print('Peso Neto de Kg del 2014: {} kg'.format(suma_2014_kg))

# Suma Peso de Qntl para el 2014 
suma_2014_qntl = get_sumpeso_anio(df_disac_qntl, '2014') 
print('Peso Neto de Qntl del 2014: {} qn'.format(suma_2014_qntl))

# Suma Peso Neto total en Kg para el 2014
suma_2014_total = suma_2014_kg + (suma_2014_qntl * 46.02)
print('Peso Neto Total de 2014: {} kg'.format(suma_2014_total))
```

#### Peso neto total en kg para el año 2015

```python
# Suma Peso de Kg para 2015 
suma_2015_kg = get_sumpeso_anio(df_disac_kgr, '2015')
print('Peso Neto de Kg del 2015: {} kg'.format(suma_2015_kg))

# Suma Peso de Qntl para el 2015 
suma_2015_qntl = get_sumpeso_anio(df_disac_qntl, '2015') 
print('Peso Neto de Qntl del 2015: {} qn'.format(suma_2015_qntl))

# Suma Peso Neto total en Kg para el 2015
suma_2015_total = suma_2015_kg + (suma_2015_qntl * 46.02)
print('Peso Neto Total de 2015: {} kg'.format(suma_2015_total))
```

#### Peso neto total en kg para el año 2016

```python
# Suma Peso de Kg para 2016 
suma_2016_kg = get_sumpeso_anio(df_disac_kgr, '2016')
print('Peso Neto de Kg del 2016: {} kg'.format(suma_2016_kg))

# Suma Peso de Qntl para el 2016 
suma_2016_qntl = get_sumpeso_anio(df_disac_qntl, '2016') 
print('Peso Neto de Qntl del 2016: {} qn'.format(suma_2016_qntl))

# Suma Peso Neto total en Kg para el 2016
suma_2016_total = suma_2016_kg + (suma_2016_qntl * 46.02)
print('Peso Neto Total de 2016: {} kg'.format(suma_2016_total))
```

### Gráfico lineal de evolución anual de producción en Kg

###  Suma total Peso neto por cultivos
#### Función para obtener suma total peso neto por cultivos

```python
import os
def get_disac_cultivo():
    df_disac_cultivo = pd.DataFrame(columns=["IDCULTIVO","DESCRIPCION", "PESONETO"])

    df_disac_cultivo[["IDCULTIVO", "DESCRIPCION"]]  = df_cultivos[["IDCULTIVO", "DESCRIPCION"]] 

    df_disac["IDCULTIVO"].to_csv("df_disac.csv")
    
    for i in range (0, len(df_disac_cultivo)):
        os.system("grep " + ','+ df_cultivos["IDCULTIVO"][i] + " df_disac.csv > df_cultivos_repeat.csv")
        size_cultivos = os.path.getsize('df_cultivos_repeat.csv')

        if size_cultivos > 0 :
            #df_cultivos_temp = pd.read_csv("df_cultivos_repeat.csv")
            df_disac_sum = df_disac.groupby(['IDCULTIVO']).get_group(df_disac_cultivo["IDCULTIVO"][i])["PESONETO"].sum()
            if df_disac_cultivo["IDCULTIVO"][i] == '0008':
                #lleno peso neto si es quintales, pasando a kg
                df_disac_cultivo["PESONETO"][i] = df_disac_sum * 46.02
            else:
                #lleno peso neto kg
                df_disac_cultivo["PESONETO"][i] = df_disac_sum
        else:
            # Asigno 0 si no encuentra idcultivo de df_cultivos en df_disac
            df_disac_cultivo["PESONETO"][i] = 0
    return df_disac_cultivo
```

```python
# Invocanco a la función para obtenr suma total peso neto por cultivos
df_peso_cultivo = get_disac_cultivo()
df_peso_cultivo
```

```python
# Función para obtener suma total anual en Kg por cultivos
import os
def get_disac_cultivo_a(df_a ):
    df_disac_cultivo = pd.DataFrame(columns=["IDCULTIVO","DESCRIPCION", "PESONETO"])

    df_disac_cultivo[["IDCULTIVO", "DESCRIPCION"]]  = df_cultivos[["IDCULTIVO", "DESCRIPCION"]] 
    df_a["IDCULTIVO"].to_csv("df_disac.csv")
    
    for i in range (0, len(df_disac_cultivo)):
        os.system("grep " + ','+ df_cultivos["IDCULTIVO"][i] + " df_disac.csv > df_cultivos_repeat.csv")
        size_cultivos = os.path.getsize('df_cultivos_repeat.csv')
        #print ('position: {}; idcultivo: {}; size_cultivos: {} '.format(i,df_cultivos["IDCULTIVO"][i], size_cultivos))
        if size_cultivos > 0 :
            
            df_disac_sum = df_a.groupby(['IDCULTIVO']).get_group(df_disac_cultivo["IDCULTIVO"][i])["PESONETO"].sum()
            if df_disac_cultivo["IDCULTIVO"][i] == '0008':
                df_disac_cultivo["PESONETO"][i] = df_disac_sum * 46.02
            else:
                df_disac_cultivo["PESONETO"][i] = df_disac_sum
        else:
            df_disac_cultivo["PESONETO"][i] = 0
    return df_disac_cultivo
```
